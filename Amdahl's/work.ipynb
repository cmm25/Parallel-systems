```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
processors = [10,100,1000,10000,100000]
p_values = [0.50, 0.90, 0.95, 0.99]
color_scheme = {0.50:"red", 0.90:'green', 0.95:'orange',0.99:'blue'}
```

```python
def calculate_speedup(p, n):
    return 1 / ((p / n) + (1 - p))
```

```python
speedup_values = {}
for p in p_values:
    speedup_values[p] = [calculate_speedup(p, n) for n in processors]
```

```python
print("Speedup Values:")
for p, speedups in speedup_values.items():
    print(f"p = {p}:")
    for n, speedup in zip(processors, speedups):
        print(f"  Processors: {n}, Speedup: {speedup:.3f}")
    print()
```

```python
plt.figure(figsize=(15,10))

for f in speedup_values:
    plt.semilogx(processors, speedup_values[f], color=color_scheme[f], 
                    label=f'f={f}', marker='o')

plt.grid(True)
plt.xlabel('Number of Processors')
plt.ylabel('Speedup')
plt.title("Amdahl's Law: Speedup vs Number of Processors")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
parallel = [0.6,0.9,0.99]
serial = [0.4,0.10,0.1]
n_processors =[10,10,10]

```

```python
def speedup(parallel, serial, n_processors):
    speedup_values = []
    for p, s, n in zip(parallel, serial, n_processors):
        task_speedup = 1 / ((p / n) + s)
        speedup_values.append(task_speedup)
    return speedup_values
```

```python
results = speedup(parallel, serial, n_processors)
print("Speedup values:", results)
```
